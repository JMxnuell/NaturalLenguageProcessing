***
* Función para el BOW:


```
def BOW(dct,text):
    """Función que contabilizará la frecuencia de palabras encontradas en el texto.
       Parámetros: dct -> defaultdict(int), text -> string """

    for token in re.split(signos,text.lower()): # separamos las palabras por signos de puntuación o espacios
        if token != '': #si no es una palabra vacía
            dct[token] += 1 # aumentamos la frecuencia del token
```

---


```
d1 = defaultdict(int)
d2 = defaultdict(int) # Diccionario para contabilizar frecuencia de palabras
d3 = defaultdict(int)
BOW(d1, text1)
BOW(d2, text2)
BOW(d3, text3)
list_bow = [d1,d2,d3]
df = pd.DataFrame(list_bow)
df.index = ["texto1", "texto2", "texto3"] 
df.fillna(0) # rellenamos con 0 las palabras que no se encontraron en el texto
```

```
import numpy as np
vec1 = df.loc["texto1"].values
vec2 = df.loc["texto2"].values
vec3 = df.loc["texto3"].values
vec1[np.isnan(vec1)] = 0
vec2[np.isnan(vec2)] = 0
vec3[np.isnan(vec3)] = 0
```

* Función similitud coseno:

```
import math
def cos_sim(v1, v2):
    """Función similitud coseno que devuelve la similitud entre dos vectores"""
    pPunto = np.sum(v1*v2)
    normaA = math.sqrt(np.sum(v1*v1))
    normaB = math.sqrt(np.sum(v2*v2))
    return pPunto/(normaA*normaB)
```
