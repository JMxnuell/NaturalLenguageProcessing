```
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```
columns = ['Pisos', 'Estructura', 'Uso', 'Propiedad', 'PisosSubterr']
pisos = [3, 3, 3]
estructura = [['semi-abierta', 'cerrada', 'cerrada'],[0,1,1]]
uso = [['escolar', 'escolar', 'escolar'], [1,1,1]]
propiedad = [['public', 'public', 'public'], [1,1,1]]
subterr = [['si', 'si',' no'], [1,1,0]]
df = pd.DataFrame(list(zip(pisos, estructura[0], uso[0], propiedad[0], subterr[0])), index=['ESCOM', 'CIC', 'Biblioteca'])
df2 = pd.DataFrame(list(zip(pisos, estructura[1], uso[1], propiedad[1], subterr[1])), index=['ESCOM', 'CIC', 'Biblioteca'])
df.columns = columns
df2. columns = columns
```

```
cosine_similarity(df2.loc["ESCOM":"ESCOM"],df2.loc["CIC":"Biblioteca"])
```

```
cosine_similarity(df2.loc["Biblioteca":"Biblioteca"],df2.loc["CIC":"CIC"])
```

```
import math
def distance(df, v1, v2):
    sum = 0;
    for col in df:
        sum += (df.loc[v1][col] - df.loc[v2][col])**2
    return math.sqrt(sum)
```

```
distance(df2,'ESCOM','CIC')
```

```
distance(df2,'ESCOM','Biblioteca')
```

```
distance(df2,'Biblioteca','CIC')
```
